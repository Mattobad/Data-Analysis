```python
# standard libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

###### Trend in the USA 

```python
data_usa = pd.read_csv('data/data_trends_usa.csv',skiprows=2)
data_usa.head()
```

```python
#check for missing values
data_usa.isnull().sum()
```

```python
# fix column names
col_names = ['week','data_science','machine_learning','deep_learning',
            'data_analytics']
data_usa.columns = col_names
data_usa.tail()
```

```python
# convert to datetime
data_usa['week'] = pd.to_datetime(data_usa['week'])
```

```python
# set the week as index
data_usa.set_index('week',inplace=True)
data_usa.head(3)
```

```python
data_usa.describe()
```

```python
# plot the time series of google searches of data science trends
plt.style.use('seaborn-deep')
ax = data_usa.plot(color=['red','green','blue','orange'],
                   grid=True,linewidth=2,alpha=0.5)
ax.set_xlabel('Month')
ax.set_ylabel('Number of Searches')
ax.set_title('Google Trends in data science fields')
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), 
           loc=2, borderaxespad=0.)
plt.show()
```

Visually inspecting the time series above, we can see that all the data science related field have stable from the months end of September to middle of November. These are followed either by down or stable trends expect in the between the months Nov-Dec, Dec-Jan and Mar-Apr where you can see notable decrease in the number of search.

```python
# check options for fonts, lines, styles
print(plt.style.available)
```

```python
# plot histogram(frequency of counts), change num of bins to see 
# different plots
data_usa.plot(kind='hist',bins=25,
              color=['red','green','blue','orange'],
              grid=True,alpha=0.3)
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), 
           loc=2, borderaxespad=0.)
plt.show()
```

A density plot shows the distribution of the data over a continuous interval. Kernel density plot smoothes out the noise in the time series data. The peaks of a density plot help display where values are concentrated over the interval. A kernel density plot is a better way to display the distrubtion because it's not affected by the number of bins used( each bar used in a typical histogram).

```python
data_usa.plot(kind='density',color=['red','green','blue','orange'],
              grid=True, linewidth=3,
            fontsize=10,alpha=0.5)
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), 
           loc=2, borderaxespad=0.)
plt.show()
```
